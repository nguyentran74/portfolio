### **Studio Class (SC) 15, Question 1:** <a id="q15a"></a>

List 3 benefits of normalisation

***Answer:***

- *Prevents Redundancy: It ensures that identical data isn't stored in multiple locations.*
- *Maintains Data Uniformity: It ensures the accuracy and consistency of data throughout the database.*
- *Streamlines Modifications: It simplifies the process of making updates, additions, and deletions in the database.*

### **Studio Class (SC) 15, Question 2:** <a id="q15b"></a>

Activity - Describe in one sentence (in your own words)

|Question | Answer |
|---|---|
| 1NF | *A table meets the criteria for 1NF when every cell (field) contains a single data element, and each entry within a column is consistent in appearance.* | 
| 2NF | *A table reaches 2NF status when it is in 1NF and all its contents are directly related to the table's primary key.* |
| 3NF | *A table achieves 3NF when it is in 2NF and every piece of information in the table is independent, requiring no other data within the table to be interpreted.* |

### **Studio Class (SC) 15, Question 6:** <a id="q15f"></a>

Identify and explain if the Bus entity satisfies:
1st Normal Form
2nd Normal Form
3rd Normal Form

|Question | Answer |
|---|---|
| 1NF | *The Bus entity meets the criteria for 1NF as every bus is identified by a distinct registration number (Rego), and attributes such as Brand, Model, YearMade, noSeats, ServiceInterval, and LastServiceDate each contain a single, indivisible value per record, guaranteeing that the data entries are atomic.* | 
| 2NF | *The Bus entity conforms to 2NF because it has achieved 1NF, and every attribute is completely reliant on the primary key (Rego) without partial dependence, ensuring that no attribute depends solely on a portion of the primary key.* |
| 3NF | *The Bus entity adheres to 3NF standards as it is already in 2NF and lacks any transitive dependencies. This means that attributes like Brand, Model, YearMade, and so on, rely directly on the primary key (Rego) rather than on any other non-key attributes.* |

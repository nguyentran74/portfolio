### **Studio Class (SC) 13, Question 2:** <a id="q13b"></a>

What is the essential difference between a WHERE clause and the HAVING clause?

***Answer:*** *The fundamental distinction between a WHERE clause and a HAVING clause lies in their respective purposes. The WHERE clause is employed to filter rows before they undergo any grouping, typically preceding the GROUP BY clause. It operates on individual rows within the dataset. Conversely, the HAVING clause serves the purpose of filtering groups that have already been formed as a result of the GROUP BY clause. It comes into play after the grouping operation, working on aggregated data.*

### **Studio Class (SC) 13, Question 3:** <a id="q13c"></a>

Explain what a trigger is and why we use them.

***Answer:***

*A database trigger functions akin to an automated alarm system that activates in response to specific events occurring within the database, such as data insertion, modification, or deletion. Triggers serve two primary purposes:*

- *Ensuring Data Accuracy: Triggers act as safeguards to ensure that data entered into the database adheres to predefined rules and correctness standards. For instance, a trigger can validate whether the price of a product falls within an acceptable range when it is being added to the database.*

- *Automation of Tasks: Triggers operate autonomously to perform certain actions when changes take place within the database. For instance, when one part of the data is updated, a trigger can be configured to automatically update another related part, streamlining data management and maintenance processes.*

### **Studio Class (SC) 13, Question 5:** <a id="q13e"></a>

Explain two advantages of using a view.

***Answer:*** 

*The utilization of views in a database offers two noteworthy advantages:*
- *Enhanced Privacy: Views have the capability to conceal specific portions of the data, allowing you to exhibit only the pertinent database information. This results in a reduction of data complexity and minimizes information overload, thereby safeguarding privacy and maintaining data confidentiality.*

- *Simplified Querying and Convenience: Views streamline the process of executing intricate queries. Rather than composing lengthy and convoluted queries repeatedly, you can create a view that encapsulates the complexity. Subsequently, you can effortlessly employ the view, benefiting from a more straightforward and convenient approach to accessing the desired data.*
